# **Face detection using HAAR Cascade classifiers**

```python
import cv2
face_classifier = cv2.CascadeClassifier("Haarcascade/haarcascade_frontalface_default.xml")

# load our image then convert it to grayscale
img = cv2.imread("Examples/pavan.jpg")
img = cv2.resize(img,(500,500))
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Tuning cascade classifier - detectmultiscale(input image, **Scale factor**, **Min neighbors**)
faces = face_classifier.detectMultiScale(gray, 1.5, 6)

#When no faces detected, face classifier returns the empty tuple
if faces is ():
    print("No faces found")

#we iterate through our faces array and draw a rectangle over each face in faces
for(x,y,l,h) in faces:
    cv2.rectangle(img,(x,y), (x+l,y+h), (0,255,100),3)

cv2.imshow("Face Detection", img)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```

# **Face and Eye detection using HAAR Cascade classifiers in image**

```python
face_classifier = cv2.CascadeClassifier("Haarcascade/haarcascade_frontalface_default.xml")
eye_classifier = cv2.CascadeClassifier("Haarcascade/haarcascade_eye.xml")

# load our image then convert it to grayscale
img = cv2.imread("Examples/pavan.jpg")
resized_img = cv2.resize(img,(650,650))
gray = cv2.cvtColor(resized_img, cv2.COLOR_BGR2GRAY)

faces = face_classifier.detectMultiScale(gray, 1.05, 5)

if faces is ():
    print("No faces found")
for(x,y,l,h) in faces:
    cv2.rectangle(resized_img,(x,y), (x+l,y+h), (0,255,100),5)
    roi_gray = gray[y:y+h, x:x+l]
    roi_color = resized_img[y:y+h, x:x+l]
    eyes = eye_classifier.detectMultiScale(roi_gray)
    for (ex,ey,el,eh) in eyes:
        cv2.rectangle(roi_color,(ex,ey),(ex+el,ey+eh),(0,255,100),3)
cv2.imshow("Face & Eye Detection", resized_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# **Face and Eye detection using HAAR Cascade classifiers in image**

```python
face_cascade = cv2.CascadeClassifier("Haarcascade/haarcascade_frontalface_default.xml")
eye_cascade = cv2.CascadeClassifier("Haarcascade/haarcascade_eye.xml")

def detect(gray, frame):
    faces = face_cascade.detectMultiScale(gray,1.3,5)
    for ( x, y, l, h) in faces:
        cv2.rectangle(frame, (x,y),(x+l, y+h),(255,0,0),3)
        roi_gray = gray[y:y+h, x:x+l]
        roi_color = frame[y:y+h, x:x+l]
        eyes = eye_cascade.detectMultiScale(roi_gray,1.1,3)
        for (ex, ey, el, eh) in eyes:
            cv2.rectangle(roi_color, (ex, ey), (ex+el,ey+eh), (0,255,100), 3)
    return frame

video = cv2.VideoCapture(0)

while True:
    check, frame = video.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    canvas = detect(gray, frame)
    cv2.imshow("video",canvas)
    if cv2.waitKey(1) == ord("q"):
        break
video.release()
cv2.destroyAllWindows()
```

# **Pedistrian Detection(Full body detection)**

```python
import cv2

# Creating a body classifier
body_classifier = cv2.CascadeClassifier("Haarcascade/haarcascade_fullbody.xml")

#Initiate video capture for video file
cap = cv2.VideoCapture("Examples/walking.avi")

# loop once video is successfully loaded
while cap.isOpened():

    #Read first frame
    check, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    #Pass frame to our body classifier
    bodies = body_classifier.detectMultiScale(gray,1.2,3)

    #Extract bounding boxes for any bodies identified
    for (x,y,l,h) in bodies:
        cv2.rectangle(frame, (x,y), (x+l, y+h), (0,255,255),2)
        cv2.imshow('Pedestrains',frame)
    if cv2.waitKey(1)==ord("q"):
        break
cap.release()
cv2.destroyAllWindows()
```

# **Car detection**

```python
import time

#create body classifier
car_classifier = cv2.CascadeClassifier("Haarcascade/haarcascade_car.xml")

# Initiate video capture for video file
cap = cv2.VideoCapture("Examples/cars.avi")

#Loop once video is successfully loaded
while cap.isOpened():

    time.sleep(0.05)
    #Read first time
    check, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    #Pass frame to our body classifier
    cars = car_classifier.detectMultiScale(gray, 1.4, 2)

    #Extract bounding boxes for any bodies identified
    for (x,y,l,h) in bodies:
        cv2.rectangle(frame, (x,y), (x+l, y+h), (0,255,255), 2)
        cv2.imshow('Cars',frame)
    if cv2.waitKey(1)==ord("q"):
        break
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import time

# Load car classifier
car_classifier = cv2.CascadeClassifier("Haarcascade/haarcascade_car.xml")

# Open video file
cap = cv2.VideoCapture("Examples/cars.avi")

# Loop until video ends or is closed
while cap.isOpened():
    time.sleep(0.05)
    
    # Read frame
    check, frame = cap.read()
    if not check:  # Break if no frame is returned
        break

    # Convert to grayscale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect cars
    cars = car_classifier.detectMultiScale(gray, 1.4, 2)

    # Draw rectangles around detected cars
    for (x, y, w, h) in cars:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 255), 2)

    # Show the frame
    cv2.imshow("Cars", frame)

    # Exit on pressing 'q'
    if cv2.waitKey(1) == ord("q"):
        break

# Cleanup
cap.release()
cv2.destroyAllWindows()
```
